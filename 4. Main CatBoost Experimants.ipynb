```python
topic_LDA_medians = []
for i in range(1,4):
    topic_LDA_medians.append(np.median(train_df.iloc[ind_train, :].loc[y_train==i, tmc], axis=0))
topic_LDA_means = []
for i in range(1,4):
    topic_LDA_means.append(np.mean(train_df.iloc[ind_train, :].loc[y_train==i, tmc], axis=0))
topic_LDA_75_quantile = []
for i in range(1,4):
    topic_LDA_75_quantile.append(np.percentile(train_df.iloc[ind_train, :].loc[y_train==i, tmc], 75, axis=0))
topic_LDA_25_quantile = []
for i in range(1,4):
    topic_LDA_25_quantile.append(np.percentile(train_df.iloc[ind_train, :].loc[y_train==i, tmc], 25, axis=0))
```

```python
train_df.res = train_df.res.astype(int)
```

```python
from scipy.spatial.distance import cosine
def cos_dist_from_median(val):
    return cosine(topic_LDA_medians[val.res-1], val[tmc.tolist()].tolist())

def cos_dist_from_mean(val):
    return cosine(topic_LDA_means[val.res-1], val[tmc.tolist()].tolist())

def cos_dist_from_75_per(val):
    return cosine(topic_LDA_75_quantile[val.res-1], val[tmc.tolist()].tolist())

def cos_dist_from_25_per(val):
    return cosine(topic_LDA_25_quantile[val.res-1], val[tmc.tolist()].tolist())
```

```python
cos_median_distances = train_df.apply(cos_dist_from_median, axis=1)
cos_mean_distances = train_df.apply(cos_dist_from_mean, axis=1)
cos_75_percentile_distaces = train_df.apply(cos_dist_from_75_per, axis=1)
cos_25_percentile_distaces = train_df.apply(cos_dist_from_25_per, axis=1)
train_df['cos_median_distances2'] = cos_median_distances
train_df['cos_mean_distances2'] = cos_mean_distances
train_df['cos_75_percentile_distaces'] = cos_75_percentile_distaces
train_df['cos_25_percentile_distaces'] = cos_25_percentile_distaces
```

```python
train_df['prediction'] = train_df.loc[:, ['prob_bad', 'prob_super', 'prob_ok']].idxmax(axis=1).map({'prob_bad':1, 'prob_super':2, 'prob_ok':3})
```
